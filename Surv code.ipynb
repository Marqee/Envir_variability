# Knowledge of Climate change

```python
#Knowlwdge on climate change
knowledge = df.iloc[:,5:9]
knowledge = pd.concat([knowledge,df['Age_Group']],axis=1)
```

```python
knowledge.isnull().sum()
```

```python
knowledge[knowledge['Purch_Green_plastic_helps_Envir'].isnull()]
```

```python
def fillnans(df,col,gen):
    if gen == 'Gen X':
        df.loc[(df['Age_Group']==gen) & df[col].isnull(),col] = df[df['Age_Group']==gen][col].mode()[0]
        return df
    if gen == 'Gen Z':
        df.loc[(df['Age_Group']==gen) & df[col].isnull(),col] = df[df['Age_Group']==gen][col].mode()[0]
        return df
```

```python
fillnans(knowledge,col='Purch_Green_plastic_helps_Envir',gen='Gen X')
fillnans(knowledge,col='Purch_Green_plastic_helps_Envir',gen='Gen Z')
fillnans(knowledge,'Climate_change_positive_for_Summers','Gen Z')
```

```python
knowledge.isnull().sum()
```

```python
#Plot Normalized data on column
order = ['Strongly Disagree','Disagree','Agree', 'Strongly agree']

def norm_plot(df,col_name,gen,order=order,palette='Spectral',ax=None):
    reduced_df = df[[col_name,'Age_Group']]
    
    #Get percentage per col
    data_plot = reduced_df[reduced_df['Age_Group']==gen][col_name].value_counts(normalize=True)
    
    sns.barplot(y=data_plot.index,
                x=data_plot.values,
                order = order,
                ax = ax,
                palette = palette,
                edgecolor = 'black').set_title('Age {}: {}'.format(gen, col_name))
```

```python
know_col = knowledge.columns[:-1]

ages = [age for age in knowledge['Age_Group'].unique()]

for know in know_col:
    fig,ax = plt.subplots(nrows=1,ncols=2, figsize=(12,2),sharey=True, sharex=True)
    
    for a, gen in zip(ax.flatten(), ages):
        norm_plot(knowledge,know,gen,ax=a,palette='Greens')
        a.xaxis.set_tick_params(which='both',labelbottom=True)
        
        a.yaxis.label.set_visible(False)
        plt.tight_layout()
```

```python
col_list = knowledge.columns[:-1]
for x in col_list:
    print(x)
    print(knowledge[x].unique())
```

```python
num_know = pd.DataFrame(knowledge['Age_Group'])
def label_eco(df,column):
    num_know[column] = df[column].map({'Strongly agree': 4, 'Agree':3, 'Disagree':2,'Strongly Disagree':1})
    return num_know
def label_eco_reversed_know(df,column):
    num_know[column] = df[column].map({'Strongly agree': 1, 'Agree':2, 'Disagree':3,'Strongly Disagree':4})
    return num_know
```

```python
knowledge.head()
```

```python
label_eco(knowledge,'Air_poll_affecting_life_quality')
label_eco(knowledge,'Purch_Green_plastic_helps_Envir')
label_eco(knowledge,'Awareness_can_Minimize')
label_eco_reversed_know(knowledge,'Climate_change_positive_for_Summers')
```

```python
num_know.describe()
```

```python
labels = list(num_know.columns[1:])

#Means for Age Groups
means_genx = list(num_know[num_know['Age_Group']== 'Gen X'].describe().loc['mean'].values)
means_genz = list(num_know[num_know['Age_Group']== 'Gen Z'].describe().loc['mean'].values)

#Standard Deviations for Groups
std_genx = list(num_know[num_know['Age_Group']== 'Gen X'].describe().loc['std'].values)
std_genz = list(num_know[num_know['Age_Group']== 'Gen Z'].describe().loc['std'].values)


x_pos = np.arange(len(labels))
```

```python
#Plot
fig,ax=plt.subplots(figsize=(9,5))

#Data on errorbar not to overlap
trans1=Affine2D().translate(0.0,-0.1) + ax.transData
trans2=Affine2D().translate(0.0,+0.1) + ax.transData
#Plot data
ax.errorbar(x=means_genx,y=labels,fmt="o",xerr=std_genx,
            elinewidth=0.5,
            capsize=4,
            marker="o",
            ms=7,
            label='Gen X',
            transform=trans1
        )
ax.errorbar(x=means_genz,y=labels,fmt="o",xerr=std_genz,
            elinewidth=0.5,
            capsize=4,
            marker="^",
            ms=7,
            label='Gen Z',
            transform=trans2
            
        )
#plt.errorbar()
ax.set_xlim(1, 5)
ax.set_yticks(x_pos)
ax.set_yticklabels(labels,fontdict={'fontsize':14})
plt.legend(title='Generations')

plt.title('Compare Averages for Generations(Climate Change Knowledge)',fontsize=16,fontdict={'family': 'serif'})
```
